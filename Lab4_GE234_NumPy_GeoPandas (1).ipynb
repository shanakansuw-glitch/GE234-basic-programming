## 🔹 ตัวอย่างที่ 1: ใช้ NumPy คำนวณข้อมูล NDVI จากภาพ Raster


```python

import numpy as np

# สร้างข้อมูลตัวอย่างสำหรับภาพ NDVI (Normalized Difference Vegetation Index)
nir = np.array([[0.7, 0.8, 0.6], [0.9, 0.5, 0.3], [0.4, 0.7, 0.2]])
red = np.array([[0.3, 0.4, 0.2], [0.5, 0.3, 0.1], [0.2, 0.3, 0.1]])

# คำนวณค่า NDVI
ndvi = (nir - red) / (nir + red)

print("ค่า NDVI:")
print(ndvi)
```


## 🔹 ตัวอย่างที่ 2: ใช้ NumPy คำนวณค่าเฉลี่ยและค่ามากสุดของ NDVI


```python

print(f"ค่าเฉลี่ย NDVI: {np.mean(ndvi):.2f}")
print(f"ค่า NDVI สูงสุด: {np.max(ndvi):.2f}")
print(f"ค่า NDVI ต่ำสุด: {np.min(ndvi):.2f}")
```


## 🔹 ตัวอย่างที่ 3: ใช้ GeoPandas โหลดและวิเคราะห์ข้อมูล Shapefile


```python

import geopandas as gpd

# โหลดข้อมูลชั้นข้อมูลจังหวัดของประเทศไทย
gdf = gpd.read_file("https://data.opendevelopmentmekong.net/th/dataset/8f3fa1b8-cb5c-48c8-9fd7-d3c213ea23db/resource/1559cee4-fedc-4330-be9c-d8cf3dd75015/download/tha_admbnda_adm1_rtsd_20190221.zip")

# แสดงข้อมูล 5 แถว
print(gdf.head())
```


## 🔹 ตัวอย่างที่ 4: คำนวณพื้นที่ของแต่ละจังหวัด


```python
# ตรวจสอบค่า CRS (Coordinate Reference System)
print(gdf.crs)

# Reproject to a projected CRS for accurate area calculation (e.g., UTM Zone 47N for parts of Thailand)
gdf_projected = gdf.to_crs(epsg=32647)

# คำนวณพื้นที่ของแต่ละจังหวัด (หน่วยเป็นตารางเมตร) จากข้อมูลที่ถูกฉายภาพ
gdf["area_sqkm"] = gdf_projected.geometry.area / 1e6

# แสดงพื้นที่จังหวัด 5 อันดับแรกที่ใหญ่ที่สุด
print(gdf.nlargest(5, "area_sqkm")[["ADM1_EN", "area_sqkm"]])
```


## 🔹 ตัวอย่างที่ 5: ใช้ GeoPandas ทำ Spatial Join


```python

# โหลดข้อมูลชั้นข้อมูลอำเภอ
gdf_districts = gpd.read_file("https://data.opendevelopmentmekong.net/th/dataset/8f3fa1b8-cb5c-48c8-9fd7-d3c213ea23db/resource/1559cee4-fedc-4330-be9c-d8cf3dd75015/download/tha_admbnda_adm1_rtsd_20190221.zip")

# ทำ Spatial Join ระหว่างอำเภอกับจังหวัด
gdf_joined = gpd.sjoin(gdf_districts, gdf, how="inner", predicate="within")

# แสดงตัวอย่างข้อมูลที่เชื่อมโยงกัน
print(gdf_joined.head())
```

แบบฝึกหัด 1: ใช้ NumPy คำนวณค่า Mean, Max, Min ของค่า NDVI **ในอาร์เรย์ที่สร้างขึ้นเอง** **bold text**

```python
# แบบฝึกหัด 1
print(f"ค่าเฉลี่ย NDVI: {np.mean(ndvi):.2f}")
print(f"ค่า NDVI สูงสุด: {np.max(ndvi):.2f}")
print(f"ค่า NDVI ต่ำสุด: {np.min(ndvi):.2f}")
```

**แบบฝึกหัด 2**: ใช้ GeoPandas โหลด Shapefile ของจังหวัด และคำนวณพื้นที่ของแต่ละจังหวัด

```python
import geopandas as gpd

# โหลดข้อมูลชั้นข้อมูลจังหวัดของประเทศไทย
gdf_exercise2 = gpd.read_file("https://data.opendevelopmentmekong.net/th/dataset/8f3fa1b8-cb5c-48c8-9fd7-d3c213ea23db/resource/1559cee4-fedc-4330-be9c-d8cf3dd75015/download/tha_admbnda_adm1_rtsd_20190221.zip")

# Reproject to a projected CRS for accurate area calculation (e.g., UTM Zone 47N for parts of Thailand)
gdf_exercise2_projected = gdf_exercise2.to_crs(epsg=32647)

# คำนวณพื้นที่ของแต่ละจังหวัด (หน่วยเป็นตารางเมตร) จากข้อมูลที่ถูกฉายภาพ
gdf_exercise2["area_sqkm"] = gdf_exercise2_projected.geometry.area / 1e6

# แสดงข้อมูล 5 แถวแรกพร้อมพื้นที่
print(gdf_exercise2[["ADM1_EN", "area_sqkm"]].head())
```

```python
แบบฝึกหัด 3: ใช้ GeoPandas ทำ Spatial Join ระหว่างข้อมูลจังหวัดและอำเภอของประเทศกัมพูชา
```

```python
import geopandas as gpd
from IPython.display import display

gdf = gpd.read_file("https://data.opendevelopmentcambodia.net/dataset/b0609661-5c22-41ee-a569-848648620136/resource/861ce9cf-2349-404c-beb4-7421326f0007/download/district-whre-wash-implemented.zip")

print("ข้อมูลดิบ (gdf.head()):")
display(gdf.head())

gdf_districts = gpd.read_file("https://data.opendevelopmentcambodia.net/dataset/b0609661-5c22-41ee-a569-848648620136/resource/861ce9cf-2349-404c-beb4-7421326f0007/download/district-whre-wash-implemented.zip")

gdf_joined = gpd.sjoin(gdf_districts, gdf, how="inner", predicate="within")

print("ข้อมูลที่ผ่าน Spatial Join (gdf_joined ทั้งหมด):")
display(gdf_joined)
```

แบบฝึกหัด 4: ใช้ NumPy และ GeoPandas ร่วมกันเพื่อหาข้อมูลจังหวัดที่มี NDVI เฉลี่ยสูงสุด

```python
# ตรวจสอบว่า gdf_exercise2 และ ndvi ถูกสร้างขึ้นแล้ว
if 'gdf_exercise2' not in locals():
    print("gdf_exercise2 ไม่ถูกสร้างขึ้น กรุณารันแบบฝึกหัด 2 ก่อน")
    # หากยังไม่มี ให้โหลดข้อมูลใหม่
    import geopandas as gpd
    gdf_exercise2 = gpd.read_file("https://data.opendevelopmentmekong.net/th/dataset/8f3fa1b8-cb5c-48c8-9fd7-d3c213ea23db/resource/1559cee4-fedc-4330-be9c-d8cf3dd75015/download/tha_admbnda_adm1_rtsd_20190221.zip")
    gdf_exercise2_projected = gdf_exercise2.to_crs(epsg=32647)
    gdf_exercise2["area_sqkm"] = gdf_exercise2_projected.geometry.area / 1e6

if 'ndvi' not in locals():
    print("ndvi ไม่ถูกสร้างขึ้น กรุณารันตัวอย่างที่ 1 ก่อน")
    import numpy as np
    nir = np.array([[0.7, 0.8, 0.6], [0.9, 0.5, 0.3], [0.4, 0.7, 0.2]])
    red = np.array([[0.3, 0.4, 0.2], [0.5, 0.3, 0.1], [0.2, 0.3, 0.1]])
    ndvi = (nir - red) / (nir + red)

# 1. คำนวณค่าเฉลี่ยของ NDVI จาก NumPy array
mean_ndvi_value = np.mean(ndvi)
print(f"ค่าเฉลี่ย NDVI จาก NumPy array: {mean_ndvi_value:.2f}")

# 2. จำลองการกำหนดค่า NDVI เฉลี่ยให้กับแต่ละจังหวัดใน GeoDataFrame
# ในสถานการณ์จริง คุณจะต้องมีข้อมูล NDVI raster และทำการวิเคราะห์แบบ zonal statistics
# แต่ในที่นี้ เราจะใช้ค่าเฉลี่ยจาก NumPy array ที่มีเป็นตัวอย่าง
# และกระจายค่าไปตามจังหวัดต่างๆ เพื่อให้เห็นการทำงานร่วมกัน
# เพื่อให้เห็นความแตกต่าง ผมจะสุ่มค่าเล็กน้อยรอบๆ mean_ndvi_value
np.random.seed(42) # เพื่อให้ผลลัพธ์สอดคล้องกันทุกครั้งที่รัน
gdf_exercise2['simulated_avg_ndvi'] = mean_ndvi_value + (np.random.rand(len(gdf_exercise2)) - 0.5) * 0.1 # สุ่มค่า +/- 0.05 จากค่าเฉลี่ย

print("\nGeoDataFrame พร้อมค่า NDVI จำลอง (5 แถวแรก):")
display(gdf_exercise2[['ADM1_EN', 'simulated_avg_ndvi']].head())

# 3. หาจังหวัดที่มีค่า NDVI เฉลี่ยสูงสุด
province_with_max_ndvi = gdf_exercise2.loc[gdf_exercise2['simulated_avg_ndvi'].idxmax()]

print("\nจังหวัดที่มีค่า NDVI เฉลี่ยสูงสุด:")
print(f"ชื่อจังหวัด: {province_with_max_ndvi['ADM1_EN']}")
print(f"ค่า NDVI เฉลี่ย: {province_with_max_ndvi['simulated_avg_ndvi']:.2f}")
```
