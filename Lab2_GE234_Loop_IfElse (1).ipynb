## 🔹 ตัวอย่างที่ 1: การใช้ `for loop` เพื่อแสดงค่าพิกัด


```python

# พิกัดของสถานที่สำคัญในประเทศไทย
locations = {
    "Bangkok": (13.7563, 100.5018),
    "Chiang Mai": (18.7883, 98.9853),
    "Phuket": (7.8804, 98.3923),
    "Ubon Ratchathani": (15.2287, 104.8564)
}

# วนลูปแสดงค่าพิกัด
for city, coord in locations.items():
    print(f"{city}: Latitude {coord[0]}, Longitude {coord[1]}")
```


## 🔹 ตัวอย่างที่ 2: การใช้ `if-else` กรองข้อมูลพิกัด


```python

for city, coord in locations.items():
    if coord[0] > 16:
        print(f"{city} อยู่ทางภาคเหนือของประเทศไทย")
    else:
        print(f"{city} ไม่ใช่จังหวัดทางภาคเหนือ")
```


## 🔹 ตัวอย่างที่ 3: ใช้ `while loop` นับจำนวนจุดข้อมูล


```python

points = [(13.7563, 100.5018), (18.7883, 98.9853), (7.8804, 98.3923)]
count = 0

while count < len(points):
    print(f"Point {count + 1}: {points[count]}")
    count += 1
```


## 🔹 ตัวอย่างที่ 4: วิเคราะห์ข้อมูล Raster ด้วย `for loop`


```python

import numpy as np

# จำลองข้อมูล NDVI (Normalized Difference Vegetation Index)
ndvi = np.array([
    [0.2, 0.4, 0.5],
    [0.1, 0.3, 0.7],
    [0.6, 0.8, 0.2]
])

# วนลูปตรวจสอบค่าพืชพรรณ
for row in range(ndvi.shape[0]):
    for col in range(ndvi.shape[1]):
        if ndvi[row, col] > 0.3:
            print(f"พิกัด ({row}, {col}) มีพืชพรรณหนาแน่น (NDVI={ndvi[row, col]})")
```


# 📝 **กิจกรรมในแลป**

1. **แบบฝึกหัด 1**: เขียน `for loop` เพื่อแสดงชื่อจังหวัดและพิกัดจากลิสต์ของคุณเอง
2. **แบบฝึกหัด 2**: ใช้ `if-else` เพื่อแยกแยะพื้นที่ที่เป็นภูเขาและพื้นที่ราบจากลิสต์ของพิกัด
3. **แบบฝึกหัด 3**: ใช้ `while loop` เพื่อวนลูปและแสดงค่าละติจูดจากลิสต์จนกว่าจะพบค่าต่ำกว่า 10°
4. **แบบฝึกหัด 4**: ใช้ `for loop` วิเคราะห์ค่า NDVI และหาค่าพื้นที่ที่มีสภาพแห้งแล้ง (NDVI < 0.2)


```python
# สร้างลิสต์ของจังหวัดและพิกัด
my_locations = {
    "Lampang": (18.2917, 99.4975),
    "Nonthaburi": (13.8647, 100.5222),
    "Khon Kaen": (16.4328, 102.8252),
    "Krabi": (8.0776, 98.9056)
}

# วนลูปเพื่อแสดงชื่อจังหวัดและพิกัด
for city, coords in my_locations.items():
    print(f"จังหวัด: {city}, ละติจูด: {coords[0]}, ลองจิจูด: {coords[1]}")
```

```python
# ใช้ if-else เพื่อแยกแยะพื้นที่ที่เป็นภูเขาและพื้นที่ราบ
# ให้ละติจูดที่สูงกว่า 15 องศาหมายถึงพื้นที่ที่มีลักษณะภูเขาในประเทศไทย

print("การจำแนกพื้นที่ภูเขาและพื้นที่ราบจากพิกัด:")
for city, coords in my_locations.items():
    latitude = coords[0]
    if latitude > 15.0:
        print(f"{city} (ละติจูด {latitude}): เป็นพื้นที่ภูเขา")
    else:
        print(f"{city} (ละติจูด {latitude}): เป็นพื้นที่ราบ")
```

```python
# สร้างลิสต์พิกัดสำหรับแบบฝึกหัดนี้
geo_coords = [
    (18.2917, 99.4975),  # Lampang
    (13.8647, 100.5222), # Nonthaburi
    (16.4328, 102.8252), # Khon Kaen
    (8.0776, 98.9056),   # Krabi
    (10.0000, 100.0000)  # จุดสมมติสำหรับทดสอบค่าต่ำกว่า 10 (หากยังไม่พบ)
]

i = 0 # กำหนดตัวแปรสำหรับวนลูป

print("แสดงค่าละติจูดจนกว่าจะพบค่าต่ำกว่า 10 องศา:")
while i < len(geo_coords):
    latitude = geo_coords[i][0]
    print(f"ละติจูดที่พิกัดที่ {i+1}: {latitude}")
    if latitude < 10:
        print(f"พบค่าละติจูดที่ต่ำกว่า 10 องศาแล้ว ({latitude}) หยุดวนลูป")
        break # หยุดการวนลูปเมื่อพบเงื่อนไข
    i += 1
```

```python
import numpy as np

# ใช้ข้อมูล NDVI ที่มีอยู่ หรือสร้างขึ้นมาใหม่หากยังไม่มี
ndvi = np.array([
    [0.1, 0.4, 0.05],
    [0.15, 0.3, 0.7],
    [0.6, 0.8, 0.19]
])

print("วิเคราะห์พื้นที่แห้งแล้งจากค่า NDVI (NDVI < 0.2):")
for row in range(ndvi.shape[0]):
    for col in range(ndvi.shape[1]):
        if ndvi[row, col] < 0.2:
            print(f"พิกัด ({row}, {col}) มีสภาพแห้งแล้ง (NDVI={ndvi[row, col]})")
```
